# Matplotlib - Generate LineColors automatically

```python
import matplotlib.pyplot as plt
import numpy as np

```

## Normal

```python
x = np.linspace(0, 1, 10)

for i in range(1, 6):
    plt.plot(x, i * x + i, label=f'$y = {i}x + {i}$')
plt.legend(loc='best')
plt.show()
```

## Enumerate over specific colors

```python
x = np.linspace(0, 1, 10)

for i, color in enumerate(['red', 'black', 'blue', 'brown', 'green'], start=1):
    plt.plot(x, i * x + i, color=color, label=f'$y = {i}x + {i}$')
plt.legend(loc='best')
plt.show()
```

## Select automatically a specified number of colors from an existing colormap

```python
x = np.linspace(0, 1, 10)

number = 5
cmap = plt.get_cmap('rainbow')
colors = [cmap(i) for i in np.linspace(0, 1, number)]

for i, color in enumerate(colors, start=1):
    plt.plot(x, i * x + i, color=color, label=f'$y = {i}x + {i}$')
plt.legend(loc='best')
plt.show()
```

## Select automatically a specified number of colors from an existing colormap

```python
import matplotlib.cm as cm
```

```python
number = 5
colors = cm.rainbow(np.linspace(0, 1, number))

for i in range(1, number+1):
    plt.plot(x, i * x + i, color=colors[i-1] , label=f'$y = {i}x + {i}$')
plt.legend(loc='best')
plt.show()
```
