# Pandas plot examples

```python
import pandas as pd
import numpy as np
```

```python
index = list(range(1,101)) 
columns = list("ABC")
data = np.random.randn(100, 3)

df = pd.DataFrame(data, index=index, columns=columns)
df.head()
```

```python
df.plot()
```

```python
df["A"].plot()
```

```python
df['date'] = pd.date_range("1/1/2000", periods=100)

df.plot(x='date', y='A')
```

```python
df.plot(subplots = True, layout=(2,2), figsize=(15,10))
```

```python
index = list(range(1,101)) 
columns = list("AB")
data = np.random.randn(100, 2)

df = pd.DataFrame(data, index=index, columns=columns)
df["B"] = df["B"]+10

df.plot(secondary_y=["B"], figsize=(15,10))
df.head()
```
